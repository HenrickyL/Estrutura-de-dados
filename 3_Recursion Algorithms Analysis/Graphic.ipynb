```python
import numpy as np
import matplotlib.pyplot as plt
def CriaGraf(path,Title, qtd=2,Legenda = "data"):
    M = np.loadtxt(path)
    n = []
    I = []
    R = []
    for i in range(0,len(M[:])): n.append(M[i][0])
    for i in range(0,len(M[:])): I.append(M[i][1])
    if(qtd ==2):
        for i in range(0,len(M[:])): R.append(M[i][2])

    #minimos quadrados
    #graficL(n,I,eixox='x',eixoy='y',titulo='',L='')
    #plt.show()
    if(qtd ==2):
        plt.plot(n,I,"r:o",label = "Interativo")
        plt.plot(n,R,"b:*",label = "Recusivo")
    else: plt.plot(n,I,"b:o",label = Legenda)

    plt.title(Title,fontweight="bold")
    plt.xlabel("Tamanho do vetor")
    plt.ylabel("MicroSegundos")
    plt.legend()
    plt.show()

 
```

```python
CriaGraf("Dados/dadosBubble.txt","BubbleSort Vetor")
CriaGraf("Dados/dadosBubbleList.txt","BubbleSort Lista",1)

CriaGraf("Dados/dados_Insertion.txt","InsertionSort Vetor")
CriaGraf("Dados/dadosInsertionList.txt","InsertionSort Lista",1)

CriaGraf("Dados/dados_Seletion.txt","SelectionSort Vetor")
CriaGraf("Dados/dadosSelectionList.txt","SelectionSort Lista",1)

CriaGraf("Dados/dadosHeap.txt","HeapSort Vetor")
CriaGraf("Dados/dadosHeapList.txt","HeapSort Lista",1)

CriaGraf("Dados/dadosQuick.txt","QuickSort Vetor",1,"Recursive")
CriaGraf("Dados/dadosMerge.txt","MergeSort Vetor",1,"Recursive")

```
